**Data Cleaning**

```python
# Removing all rows with NaNs in NAICS Code
cleaned_data = data.dropna(subset=["NAICS Code"])
```

```python
cleaned_data.head()
```

```python
cleaned_data["INSURED_ST_CDE"].value_counts()
```

```python
# Some state examples are lowercased
cleaned_data["INSURED_ST_CDE"] = cleaned_data["INSURED_ST_CDE"].str.upper()
```

```python
cleaned_data["INSURED_ST_CDE"].value_counts()
```

```python
# Renaming columns for consistency
cleaned_data = cleaned_data.rename(columns={"\xa0kgupta@berkleymp.com" : "Business_Name"})
cleaned_data = cleaned_data.rename(columns={"NAICS Code" : "NAICS_Code"})
cleaned_data = cleaned_data.rename(columns={"NAICS Description" : "NAICS_Description"})
```

```python
cleaned_data.head()
```

```python
cleaned_data = cleaned_data.fillna('')

# Create the unified address column
cleaned_data['FULL_ADDRESS'] = (
    cleaned_data['INSURED_ADDR_LN_1'].str.strip() + ' ' + 
    cleaned_data['INSURED_ADDR_LN_2'].str.strip() + ', ' + 
    cleaned_data['INSURED_CTY_NM'].str.strip() + ', ' + 
    cleaned_data['INSURED_ST_CDE'].str.strip() + ' ' + 
    cleaned_data['INSURED_ZIP_CDE'].astype(str).str.strip()
)

# Clean up double commas if Address Line 2 was empty
cleaned_data['FULL_ADDRESS'] = cleaned_data['FULL_ADDRESS'].str.replace(', ,', ',').str.strip()
cleaned_data.head()
```

```python
cleaned_data = cleaned_data.drop(columns=["Index", "INSURED_ADDR_LN_1", "INSURED_ADDR_LN_2", "INSURED_CTY_NM", "INSURED_CNTY_NM", "INSURED_ST_CDE", "INSURED_ZIP_CDE"])
```

```python
cleaned_data.to_csv("data/cleaned_data.csv", index=False)
```

Prep for multi-head architecture

```python
# Add input column for single encoder
cleaned_data['inputs'] = (
        "BUSINESS: " + cleaned_data['Business_Name'].astype(str) + 
        " | ADDRESS: " + cleaned_data['FULL_ADDRESS'].astype(str) + 
        " | DESC: " + cleaned_data['NAICS_Description'].fillna("").astype(str)
    )
```

```python
# Ensure the column is string-type and padded (NAICS can sometimes lose leading zeros)
cleaned_data["NAICS_Code"] = cleaned_data["NAICS_Code"].astype(str).str.zfill(6)

# Extract the hierarchical levels
cleaned_data['sector_label'] = cleaned_data['NAICS_Code'].str[:2]    # 2-digit (e.g., 44)
cleaned_data['subsector_label'] = cleaned_data['NAICS_Code'].str[:3] # 3-digit (e.g., 441)
cleaned_data['industry_label'] = cleaned_data['NAICS_Code'].str[:6]          # 6-digit (e.g., 441320)
```

The following are the class counts used for the model arch

```python
# Count number of unique codes at each level
print(f"Unique 2-digit sectors: {cleaned_data['sector_label'].nunique()}")
print(f"Unique 3-digit subsectors: {cleaned_data['subsector_label'].nunique()}")
print(f"Unique 6-digit industries: {cleaned_data['industry_label'].nunique()}")
```

```python
from sklearn.model_selection import train_test_split

# Handle singletons (codes appearing only once)
counts = cleaned_data['NAICS_Code'].value_counts()
singletons = counts[counts == 1].index
df_multi = cleaned_data[~cleaned_data['NAICS_Code'].isin(singletons)]
df_single = cleaned_data[cleaned_data['NAICS_Code'].isin(singletons)]

# Split: Separate the 20% TEST set
# Remaining 80% will be used for Train + Val
train_val, test = train_test_split(
    df_multi, 
    test_size=0.20, 
    random_state=101, 
    stratify=df_multi['NAICS_Code']
)

# Second Split: Split the 80% into Train (70% total) and Val (10% total)
# Calculation: 0.10 / 0.80 = 0.125
train, val = train_test_split(
    train_val, 
    test_size=0.125, 
    random_state=101, 
    stratify=train_val['NAICS_Code']
)

# Add singletons to Training (They must be seen during training)
train = pd.concat([train, df_single])

print(f"Final Split Counts:\nTrain: {len(train)}\nVal: {len(val)}\nTest: {len(test)}")
```

```python
train.to_csv("data/split_data/train.csv", index=False)
val.to_csv("data/split_data/val.csv", index=False)
test.to_csv("data/split_data/test.csv", index=False)
```
