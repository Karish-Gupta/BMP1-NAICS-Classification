**Exploratory Data Analysis**

```python
import pandas as pd
import openpyxl
import numpy as np

cleaned_data = pd.read_csv("data/split_data/website_summaries_val.csv")
cleaned_data.head()
```

Prep for multi-head architecture

```python
cleaned_data["summary"] = cleaned_data["summary"].replace(
    "Insufficient content for summary.", ""
)
```

```python
cleaned_data["summary"] = cleaned_data["summary"].fillna("")

cleaned_data["inputs"] = (
    "BUSINESS: " + cleaned_data["Business_Name"] +
    " | ADDRESS: " + cleaned_data["FULL_ADDRESS"] +
    cleaned_data["summary"].apply(
        lambda x: f" | SUMMARY: {x}" if x.strip() != "" else ""
    )
)
```

```python
# Ensure the column is string-type and padded (NAICS can sometimes lose leading zeros)
cleaned_data["NAICS_Code"] = cleaned_data["NAICS_Code"].astype(str).str.zfill(6)

# Extract the hierarchical levels
cleaned_data['sector_label'] = cleaned_data['NAICS_Code'].str[:2]    # 2-digit (e.g., 44)
cleaned_data['subsector_label'] = cleaned_data['NAICS_Code'].str[:3] # 3-digit (e.g., 441)
cleaned_data['industry_label'] = cleaned_data['NAICS_Code'].str[:6]          # 6-digit (e.g., 441320)
```

```python
# Drop summary column
cleaned_data = cleaned_data.drop(columns=["summary"])
```

```python
cleaned_data.to_csv("val_data_with_summaries.csv", index=False)
```
